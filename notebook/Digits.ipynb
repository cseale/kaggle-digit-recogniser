```python
# read data from csv

import pandas as pd

digits = pd.read_csv('../../datasets/digits.csv')
digits.head()
```

```python
# create test and training sets

import numpy as np
from sklearn import cross_validation, preprocessing

X = digits.ix[:,'pixel0':'pixel783'].as_matrix()
X = preprocessing.scale(X)
y = digits.label.as_matrix()
labels = np.unique(y)

[X_train, X_test, y_train, y_test] = cross_validation.train_test_split(X, y, test_size = 0.1)

print('training and test sets created')
```

```python
# create logisitic regression model, one vs all

from sklearn import linear_model
from sklearn.metrics import accuracy_score

clf = []

for c in np.nditer(labels):
    # get labels as 1s and 0s
    y_train_clf = np.copy(y_train)
    y_train_clf = y_train_clf == c
    y_train_clf.astype(int)
    print('training classifer for: ' + str(c))
    clf.append(linear_model.LinearRegression().fit(X_train, y_train_clf))
    print('training finished')
    
print('created all classifiers')
```
